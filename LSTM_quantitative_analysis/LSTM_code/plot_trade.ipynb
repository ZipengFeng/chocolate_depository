```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

# 画交易图

```python
path = '/home/chocolate/LSTM-source/daily_datas/'
# path ='C:/Users/Administrator/Desktop/daily_datas/'
data1 = pd.read_csv(path + '20170210datas.csv')
# data1['Time'] = pd.to_datetime(data1['Time'])
price = list(data1['Price'])
time = list(data1['Time'])
label = list(data1['Direction'])
plt.figure(figsize=(25,8)) 
plt.title('Trade_records')
plt.plot(price,"b-",label = 'Price')
for j in range(len(label)):
        if label[j] == 'buy':
            plt.annotate('buy', xy=(j, price[j]),xytext=(j,price[j] + 2),arrowprops=dict(facecolor='red', shrink=0.002))
        elif label[j] == 'sell':
            plt.annotate('sell', xy=(j, price[j]),xytext=(j,price[j] - 2),arrowprops=dict(facecolor='green', shrink=0.002))
        elif label[j] == 'short':
            plt.annotate('short', xy=(j, price[j]),xytext=(j,price[j] + 2),arrowprops=dict(facecolor='green', shrink=0.002))
        elif label[j] == 'cover':
            plt.annotate('cover', xy=(j, price[j]),xytext=(j,price[j] - 2),arrowprops=dict(facecolor='red', shrink=0.002))
plt.xlabel('minute')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```

# 画预测图

### ctaLSTM_V6（基于EMA走势分为五类）的预测：当预测为平稳时，表示为3，无箭头；预测上涨时，表示为1，红色下方箭头；预测上涨回调时，表示为2，灰色上方箭头；预测下跌时，表示为4，绿色上方箭头；预测下跌回调时，表示为5，黄色上方箭头。

```python
path = '/home/chocolate/LSTM-source/daily_datas/'
# path ='C:/Users/Administrator/Desktop/daily_datas/'
data1 = pd.read_csv(path + '20170208datas.csv')
# data1['Time'] = pd.to_datetime(data1['Time'])
price = list(data1['Price'])
time = list(data1['Time'])
label = list(data1['Predict'])
plt.figure(figsize=(25,8)) 
plt.title('Predict_records')
plt.plot(price,"b-",label = 'Price')
for j in range(len(label)):
        if label[j] == 1:
            plt.annotate('1', xy=(j, price[j]),xytext=(j,price[j] - 4 ),arrowprops=dict(facecolor='red', shrink=0.002))
        elif label[j] == 2:
            plt.annotate('2', xy=(j, price[j]),xytext=(j,price[j] + 4 ),arrowprops=dict(facecolor='gray', shrink=0.002))
        elif label[j] == 4:
            plt.annotate('4', xy=(j, price[j]),xytext=(j,price[j] + 4 ),arrowprops=dict(facecolor='green', shrink=0.002))
        elif label[j] == 5:
            plt.annotate('5', xy=(j, price[j]),xytext=(j,price[j] - 4 ),arrowprops=dict(facecolor='yellow', shrink=0.002))

plt.xlabel('minute')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```

### ctaLSTM_V4（基于涨跌情况分为七类）的预测：当预测为平稳时，不表示；其他的预测结果，从7到1分别用红、粉、黄、绿、灰、黑六种颜色标出，上涨箭头画在曲线下方，下跌箭头画在曲线上方。

```python
path = '/home/chocolate/LSTM-source/daily_datas/'
# path ='C:/Users/Administrator/Desktop/daily_datas/'
data1 = pd.read_csv(path + '20170210datas.csv')
# data1['Time'] = pd.to_datetime(data1['Time'])
price = list(data1['Price'])
time = list(data1['Time'])
label = list(data1['Predict_5'])
plt.figure(figsize=(25,8)) 
plt.title('Predict_records')
plt.plot(price,"b-",label = 'Price')
for j in range(len(label)):
        if label[j] == 1:
            plt.annotate('1', xy=(j, price[j]),xytext=(j,price[j] + 2 ),arrowprops=dict(facecolor='black', shrink=0.002))
        elif label[j] == 2:
            plt.annotate('2', xy=(j, price[j]),xytext=(j,price[j] + 2 ),arrowprops=dict(facecolor='gray', shrink=0.002))
        elif label[j] == 3:
            plt.annotate('3', xy=(j, price[j]),xytext=(j,price[j] + 2 ),arrowprops=dict(facecolor='green', shrink=0.002))
        elif label[j] == 5:
            plt.annotate('5', xy=(j, price[j]),xytext=(j,price[j] - 2 ),arrowprops=dict(facecolor='yellow', shrink=0.002))
        elif label[j] == 6:
            plt.annotate('6', xy=(j, price[j]),xytext=(j,price[j] - 2 ),arrowprops=dict(facecolor='pink', shrink=0.002))
        elif label[j] == 7:
            plt.annotate('7', xy=(j, price[j]),xytext=(j,price[j] - 2 ),arrowprops=dict(facecolor='red', shrink=0.002))

plt.xlabel('minute')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
